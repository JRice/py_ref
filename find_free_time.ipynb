You are given a list of blocked intervals when the robot is charging or under maintenance:

blocked = [
    [10, 20],
    [15, 25],
    [40, 50],
    [70, 90],
]

Merge overlapping blocked intervals, then return the free intervals.

```python
import ipytest
ipytest.autoconfig()

def merge_intervals(blocked):
    intervals = sorted(blocked, key=lambda interval: interval[0])
    merged = []

    for (start, end) in intervals:
        if not merged or start > merged[-1][1]: # Check the end time of the last merged
            merged.append([start, end])
        else:
            merged[-1][1] = max(end, merged[-1][1])

    return merged

def find_free_time(blocked, day_start, day_end):
    if day_start > day_end:
        return []

    merged = merge_intervals(blocked) # We know this is sorted by start time.
    # merged=[[10, 25], [40, 50], [70, 90]]

    cursor = day_start
    free_time = []

    for (start, end) in merged:
        if end <= day_start:
            continue # We don't care about that interval at all
        if start >= day_end:
            break # We're done for the day, regardless of remaining intervals.

        # keep our output intervals in range of the day:
        clipped_start = max(start, day_start)
        clipped_end = min(end, day_end)

        if cursor < clipped_start:
            free_time.append((cursor, clipped_start))

        cursor = clipped_end

    if cursor < day_end:
        free_time.append((cursor, day_end))

    return free_time

def test_free_time():
    blocked = [
        [10, 20],
        [15, 25],
        [40, 50],
        [70, 90],
    ]
    assert find_free_time(blocked, 0, 100) == [(0, 10), (25, 40), (50, 70), (90, 100)]

ipytest.run()
```

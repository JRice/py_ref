```python
import random

def set_mid(l, r):
    return l + ((r - l) // 2) # Could be { (l + r) // 2 }, but this ensures the numbers are smaller.

max_int = 1000
size = 500

arr = random.sample(range(max_int), size)
target = random.randint(1, max_int)
arr = sorted(arr)
left = 0
right = len(arr) - 1
mid = set_mid(left, right)

print(f"Looking for {target}..", end="")
count = 0
while left <= right:
    count += 1
    print(".", end="")
    if arr[mid] == target:
        break
    elif arr[mid] > target:
        right = mid - 1
    else:
        left = mid + 1
    mid = set_mid(left, right)

if left <= right:
    print(f"Done after {count} checks!")
    print(f"It was in position {mid}.")
    print("[", end="")
    print(",".join(map(str, arr[max(mid-4,0):mid])), end="")
    print(f",**{arr[mid]}**,", end="")
    print(",".join(map(str, arr[mid+1:mid+6])), end="")
    print("]")
else:
    print("Failed.")
    print(f"{target=} wasn't in the array:")
    print(f"{arr=}")
```

```python
import heapq
from math import ceil
from typing import TypedDict

class Robot(TypedDict):
    id: int
    speed: int

class Task(TypedDict):
    id: string
    available_at: int
    time_required: int

def release_workers(time, busy, free):
    while busy and busy[0][0] <= time:
        _, _, robot = heapq.heappop(busy)
        heapq.heappush(free, [-robot['speed'], robot['id'], robot])

def solve(robots, tasks):
    sorted_tasks = sorted(tasks, key=lambda task: task['available_at'])

    if not robots:
        raise ValueError("At least one robot must be available for work.")

    free = []
    busy = []
    time = 0
    assignments = {}

    for robot in robots:
        heapq.heappush(free, [-robot['speed'], robot['id'], robot])

    for task in sorted_tasks:
        time = task['available_at']

        release_workers(time, busy, free)

        if not free:
            time, _, robot = heapq.heappop(busy)
            heapq.heappush(free, [-robot['speed'], robot['id'], robot])
            release_workers(time, busy, free)

        _, _, robot = heapq.heappop(free)

        time_required = ceil(task['time_required'] / robot['speed'])
        finished_at = time + time_required

        assignments[task['id']] = {
            "worker_id": robot["id"],
            "start": time,
            "finish": finished_at,
        }

        heapq.heappush(busy, [finished_at, robot['id'], robot])
    
    return assignments

robots = [
    {"id": 1, "speed": 1},
    {"id": 2, "speed": 2},
]
tasks = [
    {'id':'A',
     'available_at':1,
     'time_required':2
    },
    {'id':'B',
     'available_at':2,
     'time_required':1
    },
    {'id':'C',
     'available_at':2,
     'time_required':4
    },
    {'id':'D',
     'available_at':3,
     'time_required':5
    },
    {'id':'E',
     'available_at':4,
     'time_required':2
    },
    {'id':'F',
     'available_at':4,
     'time_required':2
    },
]

solve(robots, tasks)
```

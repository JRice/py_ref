```python
import heapq
class Solution:
    def run_order(self, tasks: List[List[int]]) -> List[int]:
        for i, task in enumerate(tasks):
            task.append(i)
        print(f"{tasks=}")

        order, heap = [], []
        i, time = 0, tasks[0][0]
        
        while i < len(tasks) or heap:
            while i < len(tasks) and tasks[i][0] <= time:
                heapq.heappush(heap, [tasks[i][1], tasks[i][2]])
                i += 1

            if not heap:
                time = tasks[i][0]
            else:
                task = heapq.heappop(heap)
                time += task[0]
                order.append(task[1])
        
        return order


solver = Solution()
order = solver.run_order([[1,2], [2,4], [3,2], [4,1]])
print(f"{order=}")
```

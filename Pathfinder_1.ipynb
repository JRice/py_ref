You are given a warehouse grid.

. means open floor
# means blocked
R is a robot start
P is a pickup location
D is a dropoff location

The robot can move up, down, left, or right. Each move costs 1. Return the minimum number of steps required for the robot to go from R to P, then from P to D. If either leg is impossible, return -1.

```python
from collections import deque
from typing import Sequence

import ipytest
ipytest.autoconfig()

def find_positions(grid):
    positions = {}
    for row_index, row in enumerate(grid):
        for col_index, cell in enumerate(row):
            if cell in {"R", "P", "D"}:
                positions[cell] = (row_index, col_index)
    return positions

def neighbors(row, col, rows, cols):
    candidates = [
        (row + 1, col),
        (row - 1, col),
        (row, col + 1),
        (row, col - 1),
    ]

    return [
        (next_row, next_col) for next_row, next_col in candidates
        if 0 <= next_row < rows and 0 <= next_col < cols
    ]

def shortest_path(grid, start, target):
    rows = len(grid)
    cols = len(grid[0])

    queue = deque([(start, 0)]) # position and distance from start
    visited = {start}

    while queue:
        (row, col), distance = queue.popleft()

        if (row, col) == target:
            return distance

        for neighbor in neighbors(row, col, rows, cols):
            if neighbor in visited:
                continue
            r, c = neighbor
            if grid[r][c] == '#':
                continue # Shall not pass
            visited.add(neighbor)
            queue.append((neighbor, distance + 1))
    
    return -1 # No path found.

def min_delivery_steps(grid):
    if not grid:
        raise ValueError("No grid, nothing to do.")

    positions = find_positions(grid)

    required = ["R", "P", "D"]
    if any(symbol not in positions for symbol in required):
        raise ValueError("grid does not include at least one each of R, P, D: invalid task.")

    robot = positions["R"]
    pickup = positions["P"]
    dropoff = positions["D"]

    to_pickup = shortest_path(grid, robot, pickup)
    if to_pickup == -1:
        return -1

    to_dropoff = shortest_path(grid, pickup, dropoff)
    if to_dropoff == -1:
        return -1

    return to_pickup + to_dropoff



def test_simple_grid():
    grid = [
        "R..D",
        ".#.#",
        ".P..",
    ]
    assert min_delivery_steps(grid) == 7

ipytest.run()
```
